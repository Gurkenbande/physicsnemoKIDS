```python
import h5py
import os

path = r"C:\Users\david\PythonProjekte\physicsnemoKIDS\examples\weather\ProjektModelle\David\Data\RainNet_mini"

files = [f for f in os.listdir(path) if f.endswith(".h5") or f.endswith(".hdf5")]
files
```

```python
f = h5py.File(os.path.join(path, files[0]), "r")
list(f.keys())
```

```python
print(f["highres"].shape)
print(f["lowres"].shape)
```

```python
import matplotlib.pyplot as plt

t = 103
hr = f["highres"][t]
lr = f["lowres"][t]

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("HR (raw)")
plt.imshow(hr, cmap="magma")
plt.colorbar()

plt.subplot(1,2,2)
plt.title("LR (raw)")
plt.imshow(lr, cmap="magma")
plt.colorbar()

plt.show()
```

```python
hr_min, hr_max = hr.min(), hr.max()
lr_min, lr_max = lr.min(), lr.max()
hr_min, hr_max, lr_min, lr_max
```
